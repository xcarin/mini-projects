Natural Language Processing (NLTK)

Sentimental Analysis

```python
# Import sentiment analysis
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

# Download nltk
nltk.download('punkt')
nltk.download('vader_lexicon')
```

```python
# Check random review 1000
text = df['review'][1000]
text
```

```python
# Split the sentence into words
tokens = nltk.word_tokenize(text)
tokens[:10]
```

```python
# Use sentimental analysis
sia = SentimentIntensityAnalyzer()
```

```python
# Check random review 1000
sia.polarity_scores(text)
```

```python
# Find sentiment result for each review
sentiments_result = []
for review in df['review']:
    sentiment = sia.polarity_scores(review)
    sentiments_result.append(sentiment)
```

```python
# Display sentiments_result in pd format
x = pd.DataFrame(sentiments_result)
x
```

```python
sentiment_x = []
for sentiment in sentiments_result:
    compound = sentiment['compound']
    if compound >= 0.7:
        sentiment_x.append('Positive')
    elif compound <= -0.3:
        sentiment_x.append('Negative')
    else:
        sentiment_x.append('Neutral')
```

```python
# Add the sentiment_x to the DataFrame
df['sentiment'] = sentiment_x
```

```python
# Display merged data
df[['name', 'review', 'rating', 'sentiment']]
```

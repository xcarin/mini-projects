### Replace na

```python
# Check the number of null values under rating column
df[df['duration'].isna()]
```

It can be observed that duration column has been filled with nan values, while rating column is filled with duration.

```python
df.loc[df['show_id'] == 's5542', 'duration'] = '74 min'
df.loc[df['show_id'] == 's5795', 'duration'] = '84 min'
df.loc[df['show_id'] == 's5814', 'duration'] = '66 min'

#see the rows after replace action
df[df['show_id'].isin(['s5542','s5795','s5814'])]
```

```python
df = df.copy()
df.loc[df['show_id'] == 's5542', 'rating'] = 'PG'
df.loc[df['show_id'] == 's5795', 'rating'] = 'R'
df.loc[df['show_id'] == 's5814', 'rating'] = 'PG'

#see the rows after replace action
df[df['show_id'].isin(['s5542', 's5795', 's5814'])]
```

```python
# Check the number of null values under rating column
df[df['rating'].isna()]
```

```python
# Replace na values with ratings
df.loc[df['show_id'] == 's5989', 'rating'] = 'PG'
df.loc[df['show_id'] == 's6827', 'rating'] = 'TV-Y7'
df.loc[df['show_id'] == 's7312', 'rating'] = 'TV-Y7'
df.loc[df['show_id'] == 's7537', 'rating'] = 'R' 

# Display the replace nan values under rating column
df[df['show_id'].isin(['s5989','s6827','s7312','s7537'])]
```

```python
# Check nan after replacing duration and rating columns
df.isna().sum()
```

```python
df['director'] = df['director'].fillna('Unnamed')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
df['rating']= df['rating'].fillna(df['rating'].mode()[0])
```

```python
# Check nan after replacing duration and rating columns
df.isna().sum()
```

### replace prefix

```python
# Replace extra s
df['show_id'] = df['show_id'].str.replace('s','')

# Replace hyphen - 
df['rating'] = df['rating'].str.replace('-','')

# Display data
df 
```

### Split string

```python
# Split strings
df['cast'] = df['cast'].str.split(',').str[0]
df['listed_in'] = df['listed_in'].str.split(',').str[0]

# Display data
df
```

### Rename column

```python
# Rename column
df.rename(columns={"show_id":"id"}, inplace = True)
df.rename(columns={"cast":"actor"}, inplace = True)
df.rename(columns={"listed_in":"genre"}, inplace = True)
df.rename(columns={"description":"review"}, inplace = True)

# Display data
df
```

### Exploratory data analysis

1. Compare sepal length and sepal width

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm",
                hue="Species", data=data)
# Display plot
plt.show()
```

> Sentosa
- Setosa has an individual cluster which is separated from versicolor and virginica.

> Versicolor and Virginica
- Versicolor and Virginica are mixed in a cluster.
- Versicolor is spreaded in the middle of the sepal lengths and widths.
- Virginica has longer sepal lengths but smaller sepal widths.

2. Compare petal length and petal width

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="PetalLengthCm", y="PetalWidthCm",
                hue="Species", data=data)
# Display plot
plt.show()
```

> Setosa

> Versicolor and Virginica

### Apply Ml

1. Load dataset

```python
from sklearn import  datasets
iris=datasets.load_iris()
```

2. Check data

```python
iris.data
```

3. Check target values

```python
iris.target
```

4. Check labels

```python
iris.target_names
```

5. Check feature names

```python
iris.feature_names
```

6. Assign (x) data dependent and (y) target independent variables

```python
x = iris.data
y = iris.target
```

7. Split dataset into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
# test size = 0.5 represent 50% test and 50% train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5, random_state = 4)
x_train
```

7. Use KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

8. Train KNN

```python
knn.fit(x_train, y_train)
```

9. Evaluate model

```python
y_pred = knn.predict(x_test)

# Display predicted result
y_pred
```

```python
# Display actual result
y_test
```

### Check accuracy

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
